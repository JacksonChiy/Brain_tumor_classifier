```python
# 前1500张是tumor类型
for path_name in train_list:
    for i in os.listdir(path_name): #显示文件夹下面所有图像的名称
        print(i)
        read_img_path = path_name+i
        read_img = cv2.imread(read_img_path) #输入图像名称，读取图像并转化成像素值
        read_img = cv2.resize(read_img,(height,width)) #修改图像尺寸
#         read_img = filters.sobel(read_img)
#         read_img = cv2.cvtColor(read_img, cv2.COLOR_BGR2GRAY)
#         ret,read_img = cv2.threshold(read_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        train_img.append(read_img)
```
